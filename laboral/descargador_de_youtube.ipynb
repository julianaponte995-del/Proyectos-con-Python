```python
import yt_dlp
import os
from IPython.display import Image, display
```

```python
# --- CONFIGURACI√ìN DE CARPETA ---
home = os.path.expanduser("~")
ruta_descargas = os.path.join(home, "Downloads")

CARPETA_DESTINO = ruta_descargas


# --- HOOK PARA PROGRESO ---
def progreso_hook(d):
    if d['status'] == 'downloading':
        percent = d.get('_percent_str', '').strip()
        speed = d.get('_speed_str', '').strip()
        eta = d.get('_eta_str', '').strip()
        print(f"\r‚¨áÔ∏è Descargando: {percent} | ‚ö° {speed} | ‚è≥ ETA {eta}", end="")
    elif d['status'] == 'finished':
        print("\nüì¶ Descarga completada, procesando archivo...")


def descargar_youtube(url, tipo='video', carpeta=CARPETA_DESTINO):

    # Verificar carpeta
    if not os.path.exists(carpeta):
        print(f"‚ö†Ô∏è No encontr√© la carpeta '{carpeta}', guardando en la carpeta actual.")
        carpeta = "."

    ruta_salida = os.path.join(carpeta, '%(title)s.%(ext)s')

    # Opciones comunes
    ydl_opts = {
        'outtmpl': ruta_salida,
        'quiet': True,
        'no_warnings': True,
        'progress_hooks': [progreso_hook]
    }

    if tipo == 'audio':
        ydl_opts['format'] = 'bestaudio/best'
        print(f"üéµ Modo AUDIO | Carpeta: {carpeta}")
    else:
        ydl_opts['format'] = 'best'
        print(f"üé¨ Modo VIDEO | Carpeta: {carpeta}")

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info = ydl.extract_info(url, download=False)

            print(f"\nüìå T√≠tulo: {info.get('title', 'Desconocido')}")

            # --- THUMBNAIL SEGURO ---
            thumbnail = info.get('thumbnail')
            if thumbnail:
                display(Image(url=thumbnail, width=300))
            else:
                print("üñºÔ∏è Este video no tiene thumbnail disponible")

            print("\n‚è≥ Iniciando descarga...")
            ydl.download([url])

            print(f"\n‚úÖ ¬°LISTO! Archivo guardado en: {carpeta}")

    except Exception as e:
        print(f"\n‚ùå Error durante la descarga: {e}")
```

```python
# Elige: 'video' o 'audio'
link = "https://youtu.be/twlvSsIuai0?si=jmmcScb61Z-xSy5t"
descargar_youtube(link, tipo='audio')
```
